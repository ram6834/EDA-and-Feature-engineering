```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df=pd.read_excel('Data_Train.xlsx')
```

```python
train_df.head()
```

```python
test_df=pd.read_excel('Test_set.xlsx')
```

```python
test_df.head()
```

```python
final_df = pd.concat([train_df,test_df])
final_df.head()
```

```python
final_df.tail()
```

```python
final_df.info()
```

```python
## Feature engineering process

final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
#converting str to int 

final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head(2)
```

```python
# final_df['Arrival_Time'].str.split(' ').str[0]

final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
final_df['Arrival_Time'].head()
```

```python
final_df.isnull().sum()
```

```python
final_df['Arrival_hour']=final_df['Arrival_Time'].apply(lambda x : x.split(':')[0])
final_df['Arrival_minute']=final_df['Arrival_Time'].apply(lambda x : x.split(':')[1])
```

```python
final_df.head(3)
```

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.head(2)
```

```python
final_df['Arrival_hour']=final_df['Arrival_hour'].astype(int)
final_df['Arrival_minute']=final_df['Arrival_minute'].astype(int)
```

```python
final_df.info()
```

```python
final_df['Dep_hour']=final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_minute']=final_df['Dep_Time'].str.split(':').str[1]
```

```python
final_df['Dep_hour']=final_df['Dep_hour'].astype(int)
final_df['Dep_minute']=final_df['Dep_minute'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df.columns
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'NaN':5})
```

```python
final_df.head(3)
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].fillna(5)
```

```python
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['Duration']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```
