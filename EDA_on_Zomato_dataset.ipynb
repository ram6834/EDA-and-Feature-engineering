# Getting the count of the values in Country column

```python
country_names=final_df['Country'].value_counts().index
```

```python
country_names
```

```python
country_values=final_df['Country'].value_counts().values
```

```python
country_values
```

# Creating Pie chart to analyze

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.f%%')
plt.title('Top 3 Countries')
plt.show()
```

Observation : Zomato maximum records or transactions are from INDIA and then US and then UK

```python
final_df.columns
```

```python
rating=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
rating.head()
```

```python

plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=rating,hue='Rating color',palette=["white","red","orange","yellow","green","darkgreen"])
plt.title('Rating Count')
plt.show()
```

```python
sns.countplot(x='Rating color',data=rating,hue='Rating color',palette=["blue","red","orange","yellow","green","darkgreen"])
plt.title('Rating Count')
plt.show()
```
